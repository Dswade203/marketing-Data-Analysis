```python
from pytrends.request import TrendReq
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
%matplotlib
```

```python
pytrend = TrendReq()
```

```python
pytrends = TrendReq(hl='en-GB',
                   tz=360), #timezone
                   
                    
```

```python
kw_list = ["see audit", "aso audit", "app marketing agency", "seo agency", "aso agency"]
```

```python
pytrend.build_payload(kw_list,
                       cat=83,
                       timeframe='today 5-y',
                       geo='',
                       gprop='')
```

```python
trends_data = pytrend.interest_over_time()
```

```python
trends_data.head()
```

```python
trends_data.plot()
```
